```python
import matplotlib.pyplot as plt
def degree_distribution_plot(G):
    degrees = dict(G.degree())
    degree_sequence = sorted(degrees.values(),reverse=True)
    plt.hist(degree_sequence)
    plt.xlabel('Degree')
    plt.ylabel('Frequency')
    plt.xticks(np.arange(min(degree_sequence), max(degree_sequence)+1, 1.0))
    plt.yticks(np.arange(min(degree_sequence), max(degree_sequence)+1, 1.0))
    plt.title('Degree Distribution Plot')
    plt.show()

    degree_prob = nx.degree_histogram(G)
    for i in range(len(degree_prob)):
        degree_prob[i]=degree_prob[i]/len(G.edges())
    degrees = range(len(degree_prob))
    plt.plot(degrees,degree_prob)
    plt.xlabel('Degree k')
    plt.ylabel('Fraction p_k of vertices with degree k')
    plt.title('Degree Distribution Plot')
    plt.show()
degree_distribution_plot(G)
```

```python
import networkx as nx

# Read the graph from the edgelist
G = nx.read_weighted_edgelist("economic_wm3.txt")

# Check and handle negative weights
edges_to_remove = []
for edge in G.edges(data=True):
    if edge[2]['weight'] < 0:
        edges_to_remove.append((edge[0], edge[1]))

# Remove edges with negative weights
for edge in edges_to_remove:
    G.remove_edge(edge[0], edge[1])

# Check node degrees
negative_degrees = [node for node, degree in G.degree() if degree < 0]

# Remove nodes with negative degrees
for node in negative_degrees:
    G.remove_node(node)

# Ensure the graph is connected
if not nx.is_connected(G):
    # If the graph is not connected, work with the largest connected component
    largest_cc = max(nx.connected_components(G), key=len)
    G = G.subgraph(largest_cc).copy()

# Check if the graph is still not empty
if G.number_of_nodes() == 0:
    raise ValueError("Graph is empty after processing.")

# Now you can calculate radius, diameter, and periphery
print(nx.radius(G))
print(nx.diameter(G))
print(nx.periphery(G))
```

```python
import networkx as nx
from prettytable import PrettyTable

# Load your dataset or create a graph
G = nx.read_weighted_edgelist("economic_wm3.txt")

# Convert the graph to an undirected graph (important for some centrality measures)
G = G.to_undirected()

# Calculate and print top 10 nodes for each centrality measure

# Edge Betweenness Centrality
edge_betweenness = nx.edge_betweenness_centrality(G)
top_edge_betweenness = sorted(edge_betweenness.items(), key=lambda x: x[1], reverse=True)[:10]

# Eigenvector Centrality
eigenvector_centrality = nx.eigenvector_centrality(G)
top_eigenvector_centrality = sorted(eigenvector_centrality.items(), key=lambda x: x[1], reverse=True)[:10]

# Closeness Centrality
closeness_centrality = nx.closeness_centrality(G)
top_closeness_centrality = sorted(closeness_centrality.items(), key=lambda x: x[1], reverse=True)[:10]

# Degree Centrality
degree_centrality = nx.degree_centrality(G)
top_degree_centrality = sorted(degree_centrality.items(), key=lambda x: x[1], reverse=True)[:10]

# Katz Centrality
katz_centrality = nx.katz_centrality_numpy(G)
top_katz_centrality = sorted(katz_centrality.items(), key=lambda x: x[1], reverse=True)[:10]

# Create tables
edge_table = PrettyTable()
edge_table.field_names = ["Edge", "Edge Betweenness"]
node_table = PrettyTable()
node_table.field_names = [ "Eigenvector Centrality", "Closeness Centrality", "Degree Centrality", "Katz Centrality"]

# Add data to the tables
for i in range(10):
    eb_edge, eb_value = top_edge_betweenness[i]
    edge_table.add_row([eb_edge, f"{eb_edge}\n{eb_value:.4f}"])

    evc_node, evc_value = top_eigenvector_centrality[i]
    cc_node, cc_value = top_closeness_centrality[i]
    dc_node, dc_value = top_degree_centrality[i]
    katz_node, katz_value = top_katz_centrality[i]

    node_table.add_row([ f"{evc_node}\n{evc_value:.4f}", f"{cc_node}\n{cc_value:.4f}", f"{dc_node}\n{dc_value:.4f}", f"{katz_node}\n{katz_value:.4f}"])

                            # Print the tables
print("Edge Centrality:")
print(edge_table)
print("\nNode Centrality:")
print(node_table)
```

```python
def draw_communities(G, partition):
    node_groups = []
    for i in partition:
        node_groups.append(list(i))

    color_map = []
    for node in G:
        if node in node_groups[0]:
            color_map.append('yellow')
        else:
            color_map.append('green')

    nx.draw(G, node_color=color_map, with_labels=True)
    plt.show()
```

```python
from collections import defaultdict
import networkx as nx
from community import community_louvain
import matplotlib.pyplot as plt

# Read the graph from the edgelist
G = nx.read_weighted_edgelist("economic_wm3.txt")

# Check and handle negative weights
edges_to_remove = []
for edge in G.edges(data=True):
    if edge[2]['weight'] < 0:
        edges_to_remove.append((edge[0], edge[1]))

# Remove edges with negative weights
for edge in edges_to_remove:
    G.remove_edge(edge[0], edge[1])

# Check node degrees
negative_degrees = [node for node, degree in G.degree() if degree < 0]

# Remove nodes with negative degrees
for node in negative_degrees:
    G.remove_node(node)

# Ensure the graph is connected
if not nx.is_connected(G):
    # If the graph is not connected, work with the largest connected component
    largest_cc = max(nx.connected_components(G), key=len)
    G = G.subgraph(largest_cc).copy()

# Run Louvain community detection
partition = community_louvain.best_partition(G)

# Group nodes (keys) by values (community_labels)
res = defaultdict(list)
for key, val in sorted(partition.items()):
    res[val].append(key)

final_partition = [set(i) for i in res.values()]
print('Final partition:', final_partition)

# Draw the communities
draw_communities(G, final_partition)

# Calculate modularity using community_louvain module
mod = community_louvain.modularity(partition, G)
print('Modularity:', mod)
```

```python
import itertools
from networkx.algorithms import community

G=nx.read_weighted_edgelist("economic_wm3.txt")
comm = community.girvan_newman(G)
index = 0
partition=next(itertools.islice(comm, index, None))
print('No of communities =',len(partition))
for i in partition:
    print(i)
draw_communities(G,partition)
mod=community.modularity(G,partition)
print(mod)
```
